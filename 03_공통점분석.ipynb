# ✅ 공통점 (1등팀 baseline 계승)

max_point_pipeline이 1등팀 코드로부터 계승한 핵심 기법들을 정리합니다.

## 1. 동일한 피처 엔지니어링

### 기후 관련 피처
- **CDH (Cooling Degree Hours)**: 냉방도시
- **CDD (Cooling Degree Days)**: 냉방도일
- **THI (Temperature-Humidity Index)**: 온습도지수
- **WCT (Wind Chill Temperature)**: 체감온도

### 타겟 통계 피처
- `dow_hour_mean/std`: 요일-시간별 평균/표준편차
- `holiday_mean/std`: 휴일 여부별 평균/표준편차
- 기타 타겟 인코딩 피처

### 순환 피처
- 시간, 요일 등의 순환적 특성을 sin/cos 변환으로 표현

## 2. 동일한 전처리

### 휴일 설정
- `add_holiday()` 함수를 통한 건물별 세밀한 휴일 설정
- 건물별 특성을 반영한 맞춤형 휴일 정의

### 이상치 제거
- `remove_outliers()` 함수를 통한 체계적인 이상치 제거
- 건물별, 시간대별 특성을 고려한 규칙 기반 제거

### 컬럼 관리
- `DROP_COLS`: 제거할 컬럼 목록 동일
- `CAT_COLS`: 범주형 컬럼 처리 방식 동일

## 3. 동일한 모델 기법

### TabPFN 사용
- Transformer 기반 사전 학습 모델
- 소규모 테이블 데이터에 특화

### Linear Regression Stacking
- TabPFN 예측값을 Linear Regression으로 앙상블
- Cross-validation을 통한 안정적인 예측

### 건물별 개별 모델링
- 각 건물의 고유한 특성을 반영
- 건물별로 독립적인 모델 학습

## 요약

max_point_pipeline은 1등팀의 핵심 기법들을 충실히 계승하면서,
실행 시간과 복잡도를 크게 줄인 **실용적인 버전**입니다.

- ✅ 검증된 피처 엔지니어링 유지
- ✅ 효과적인 전처리 방법 동일
- ✅ 핵심 모델 기법 적용
