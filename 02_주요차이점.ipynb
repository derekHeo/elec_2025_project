# 🔍 주요 차이점

1등팀 원본 코드와 max_point_pipeline의 주요 차이점을 상세히 분석합니다.

## 비교 테이블

| 항목      | 1등팀 원본                         | max_point_pipeline         | 영향                  |
|---------|--------------------------------|----------------------------|---------------------|
| 앙상블 구조  | 5개 시드 + AutoGluon + 타입별 모델     | TabPFN Stacking 단일 모델      | 🔴 성능 차이 큼         |
| CV 폴드   | 10-fold                        | 5-fold                     | 🟡 속도 vs 안정성 트레이드오프 |
| 파일 구조   | 6개 파일 분리                       | 1개 통합 노트북                  | 🟢 관리 편의성           |
| 타겟 스케일링 | ratio_all = base_ratio - 0.005 | ratio_all = base_ratio - 0 | 🟡 미세 조정 차이         |
| 이상치 규칙  | Research 53 추가 규칙              | 기본 규칙만                     | 🟡 약간의 데이터 차이      |
| 실행 시간   | ~5-6시간 (전체 앙상블)                | ~1시간 (단일 모델)               | 🟢 실용성 증가           |

## 1. 앙상블 구조 차이

### 1등팀 원본
- 5개의 서로 다른 시드로 학습된 TabPFN 모델
- AutoGluon 모델
- 건물 타입별 TabPFN 모델
- 복잡한 가중 평균 앙상블

### max_point_pipeline
- TabPFN Stacking 단일 모델
- 건물별 개별 모델링
- 간소화된 구조

## 2. Cross-Validation 차이

- **1등팀**: 10-fold CV → 더 안정적이지만 시간 소요
- **max_point_pipeline**: 5-fold CV → 빠른 실행, 약간의 안정성 감소

## 3. 파일 구조 차이

- **1등팀**: 전처리, 모델 학습, 앙상블이 각각 분리된 6개 파일
- **max_point_pipeline**: 전체 파이프라인이 하나의 노트북으로 통합

## 4. 실행 시간 비교

- **1등팀**: 약 5-6시간 (전체 앙상블 학습)
- **max_point_pipeline**: 약 1시간 (단일 모델 학습)
- **개선**: 실용성 측면에서 5배 이상 향상
